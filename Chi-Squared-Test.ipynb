```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_path = r'C:\Users\RAGHAV\Desktop\transform_4.csv'
df = pd.read_csv(file_path)

X = df.iloc[:, :22]  # Features
y = df.iloc[:, 22]    # Target

from sklearn.feature_selection import chi2
chi2_values, _ = chi2(X, y)

feature_names = df.columns[:22]

feature_chi2 = list(zip(feature_names, chi2_values))
feature_chi2.sort(key=lambda x: x[1], reverse=True)

top_features = [x[0] for x in feature_chi2[:10]]
top_chi2_values = [x[1] for x in feature_chi2[:10]]

plt.figure(figsize=(10, 6))
plt.bar(top_features, top_chi2_values, log=True)  

plt.xlabel('Features')
plt.ylabel('Chi-squared Test Results (log scale)')
plt.title('Top 10 Features based on Chi-squared Test Results (log scale)')
plt.xticks(rotation=45, ha='right')  

plt.tight_layout()
plt.show()
```
