```python
# Select columns B through V
columns_to_check = df.columns[1:22]  # Assuming columns B through V are columns 1 through 22

# Calculate mean and standard deviation for each selected column
means = df[columns_to_check].mean()
std_devs = df[columns_to_check].std()

# Define a threshold for outliers (e.g., 3 standard deviations)
threshold = 3

# Filter out outliers for each selected column
for column in columns_to_check:
    lower_bound = means[column] - threshold * std_devs[column]
    upper_bound = means[column] + threshold * std_devs[column]
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Save the cleaned data to a new CSV file
output_file_path = r'C:\Users\RAGHAV\Desktop\cleaned_file_without_outliers.csv'
df.to_csv(output_file_path, index=False)

print("Outliers removed. Cleaned data saved to:", output_file_path)
```
